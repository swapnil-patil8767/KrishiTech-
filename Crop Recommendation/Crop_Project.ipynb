```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
import pickle
```

```python
df1=pd.read_csv('/content/Crop_recommendation.csv')
df2=pd.read_csv('/content/Crop_recommendation df1.csv')
```

```python
df1
```

```python
df1=df1.drop(columns=['Fertilizer','Link'])
```

```python
df1
```

```python
df2
```

```python
df=pd.concat([df1,df2])
df
```

```python
df
```

```python
df.info()
```

```python
x=df.drop(columns=['Crop'])
y=df['Crop']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=24)
```

```python
model=RandomForestClassifier(
    n_estimators=50,
    max_features=0.5,
    bootstrap=True,
    criterion='gini',
    random_state=24,
    n_jobs=-1
)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

```python
from joblib import dump


dump(model, 'crop_recommendation.joblib')
```

```python
with open('crop_recommendation.pkl','wb') as file:
  pickle.dump(model,file)
```
